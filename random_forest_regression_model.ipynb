```python
df = df.drop(['Car_Name'], axis=1)
```

```python
df['Current_year'] = 2021
```

```python
df['No_of_years'] = df['Current_year'] - df['Year']
```

```python
df = df.drop(['Year', 'Current_year'], axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
X=df.iloc[:, 1:]
y=df['Selling_Price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

Hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV

rf_grid= {'n_estimators': np.arange(100,1200,12),
      'max_depth': [5,30,6],
      'min_samples_split': np.arange(5,100,20),
      'min_samples_leaf' : np.arange(2,20,10),
      'max_features': ['sqrt','auto']
  }

rs_model = RandomizedSearchCV(rf,param_distributions=rf_grid, n_iter=10,cv=5,verbose= 2, random_state=42, n_jobs=1 )
```

```python
rs_model.fit(X_train,y_train)
```

```python
predictions = rs_model.predict(X_test)
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test, predictions)
```
